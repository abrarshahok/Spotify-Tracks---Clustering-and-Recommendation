## Taking useful columns

```python
columns_to_keep = [
    "danceability", "energy", "key", "loudness", "mode", 
    "speechiness", "acousticness", "instrumentalness", 
    "liveness", "valence", "tempo"
]

filtered_df = df[columns_to_keep]
```

```python
one_hot_filtered_df = pd.get_dummies(filtered_df, columns=['key'], drop_first=True, dtype=int)
```

```python
one_hot_filtered_df.head()
```

## Normalizing the Data

K-Means is distance-based, so features must be on the same scale to avoid bias towards features with larger ranges.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(one_hot_filtered_df)
```

```python
normalized_data
```

## Determine the Optimal Number of Clusters (k)

Using the Elbow Method to decide the optimal number of clusters.

```python
from sklearn.cluster import KMeans

inertia = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(normalized_data)
    inertia.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

Inference:- Should choose `k` between 4-6

## Applying k=5

```python
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(normalized_data)
```

```python
one_hot_filtered_df['cluster'] = clusters
```

## Let's create a Content-Based Recommender System

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler
```

```python
# Select relevant features for content-based filtering
features = [
    "danceability",
    "energy",
    "valence",
    "tempo",
    "acousticness",
    "instrumentalness",
]

scaler = StandardScaler()
original_df[features] = scaler.fit_transform(original_df[features])
orginal_df[features]
```

```python
def content_based_recommendations(track_id, data, features, top_n=10):
    if track_id not in data["track_id"].values:
        raise ValueError(f"Track ID '{track_id}' not found in the dataset.")
        
    track_features = data[data["track_id"] == track_id][features].values

    similarities = cosine_similarity(track_features, data[features])[0]
    
    data["similarity"] = similarities

    recommendations = (
        data[data["track_id"] != track_id]
        .sort_values(by="similarity", ascending=False)
        .head(top_n)
    )
    
    return recommendations[["track_id", "track_name", "artists", "album_name", "similarity"]]
```

```python
def content_based_recommendations_with_clustering(track_id, data, features, top_n=10):
    if track_id not in data["track_id"].values:
        raise ValueError(f"Track ID '{track_id}' not found in the dataset.")
    
    track_cluster = data[data["track_id"] == track_id]["cluster"].values[0]
    
    cluster_data = data[data["cluster"] == track_cluster]
    
    track_features = data[data["track_id"] == track_id][features].values
    
    similarities = cosine_similarity(track_features, cluster_data[features])[0]
    
    cluster_data.loc[:, "similarity"] = similarities

    
    recommendations = (
        cluster_data[cluster_data["track_id"] != track_id]
        .sort_values(by="similarity", ascending=False)
        .head(top_n)
    )
    
    return recommendations[["track_id", "track_name", "artists", "album_name", "similarity"]]
```

```python
liked_track_id = "2i6Zgt3uFaupWIcpvRORMn"
liked_song = original_df.loc[original_df['track_id'] == liked_track_id]
print("Liked Song\n")
pd.DataFrame(liked_song[["track_id", "track_name", "artists", "album_name", "similarity"]])
```

```python
top_recommendations = content_based_recommendations_with_clustering(liked_track_id, original_df, features, top_n=10)

print("Top Recommendations - Clustering:\n")
pd.DataFrame(top_recommendations)
```
